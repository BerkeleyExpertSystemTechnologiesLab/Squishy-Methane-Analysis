```python
import cv2
import numpy as np
```

```python
path = "data/MOV_2581.mp4"
capture = cv2.VideoCapture(path)
#reading first frame
_, frame1 = capture.read()
#covert to grayscale
prvs_frame = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)

#get frame dimensions
height, width = prvs_frame.shape
size = (width, height)

#create mask
hsv_mask = np.zeros_like(frame1)
#make image saturation to a maximum value
hsv_mask[..., 1] = 255
img_array = []
while True:
    #capture another frame and covert to grayscale
    success, frame2 = capture.read()
    if not success:
        print("FAILED TO CAPTURE FRAME!")
        break
    else:
        print("SUCCESS!")
    next_frame = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

    # Optical flow is now calculated
    flow = cv2.calcOpticalFlowFarneback(prvs_frame, next_frame, None, 
                                        pyr_scale=0.5, 
                                        levels=5, 
                                        winsize=20, 
                                        iterations=5, 
                                        poly_n=7, 
                                        poly_sigma=1.5, 
                                        flags=0
                                        )

    # Compute magnitude and angle of 2D vector
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])

    # Set image hue value according to the angle of optical flow
    hsv_mask[..., 0] = ang * 180 / np.pi / 2
    # Set value as per the normalized magnitude of optical flow
    hsv_mask[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
    # Convert to rgb
    rgb_representation = cv2.cvtColor(hsv_mask, cv2.COLOR_HSV2BGR)
    
    #save img to img_array
    img_array.append(rgb_representation)

    #continue on to next_frame
    previous_frame = next_frame

capture.release()
cv2.destroyAllWindows()

#about 5 minutes to run one video
```

```python
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video = cv2.VideoWriter('optical_flow_vid3.mp4', fourcc, 15, size)
for i in range(len(img_array)):
    video.write(img_array[i])
video.release()

```
